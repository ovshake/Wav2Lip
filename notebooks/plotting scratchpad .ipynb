```python
import matplotlib.pyplot as plt 
from PIL import Image
import numpy as np
import os 
import glob
```

```python
layout_dict = np.arange(15).reshape((5, 3))
layout_dict
```

```python
# images = os.listdir('/home/users/abhishekm/art-flow/fomm-relative-wav2lip-v2/drive2_vid08/')
images = ['Test_06',
			'image_10',
			'source_image_6',
			'image_8',
			'image_14',
			'image',
			'Test_03',
			'source_image_3',
			'Test_02',
			'image_11',
			'image_9',
			'source_image_4',
			'image_7',
			'Test_07']
root_dir = '/home/users/abhishekm/art-flow/fomm-relative-wav2lip-v2/drive2_vid08'
audio_name = '1_v1' 
lipsync_frames = os.path.join(root_dir, images[0], audio_name, 'lipsync','*.jpg')
print(lipsync_frames)
frames = glob.glob(lipsync_frames)
total_frames = len(frames)
total_frames
```

```python
lipsync_frames
```

```python
for idx in range(total_frames):
    fig = plt.figure(constrained_layout=True, figsize=(25, 25))
    ax_dict = fig.subplot_mosaic(
        layout_dict
    )
    plt.subplots_adjust(wspace=-0.6, hspace=0.1)
    for img_idx, image in enumerate(images):
        
        frame_path = os.path.join(root_dir, image, audio_name, 'lipsync',f'{idx:06d}.jpg')
        im = Image.open(frame_path)
        ax_dict[img_idx].imshow(im) 
    
    for k in ax_dict:
        ax_dict[k].axis('off') 
    fig.show()
    break
```
